## Dot Product Example (nut buttes sales)

```python
np.random.seed(0)

sales_amounts = np.random.randint(20, size = (5,3))
sales_amounts
```

```python
# Create weekly_sales DataFrame
weekly_sales = pd.DataFrame(sales_amounts, index=["Mon", "Tues", "Wed", "Thurs","Fri"], columns = ["Alomnd Butter", "Peanut Butter", "Cashew Butter" ])
weekly_sales
```

```python
# Create Prices Column
prices = np.array([10,8,12])
prices
```

```python
prices.shape
```

```python
# Create butter_prices Dataframe
butter_prices = pd.DataFrame(prices, index=["Price"], columns= ["Alomnd Butter", "Peanut Butter", "Cashew Butter" ])
butter_prices
```

```python
# bir önceki yazılanda sorun şu ki 3 sütun ve 1 satır bekliyor. ancak ilk oluşturulduğunda 3 sütun ve belirsiz bir satır olarak shapelediğini görüyoruz. o yüzden üstteki kod tekrar yazılacak ancak reshape edilerek 
```

```python
butter_prices = pd.DataFrame(prices.reshape(1,3), index=["Price"], columns= ["Alomnd Butter", "Peanut Butter", "Cashew Butter" ])
butter_prices
```

```python
sales_amounts.shape
```

```python
prices.shape
```

```python
Total_sales = prices.dot(sales_amounts)
```

```python
# Shapes arent alligned, lets transpose
total_sales = prices.dot(sales_amounts.T)
total_sales
```

```python
# create daily_sales
butter_prices
```

```python
weekly_sales
```

```python
daily_sales = butter_prices.dot(weekly_sales.T)
daily_sales
```

```python
weekly_sales
```

```python
weekly_sales["Total ($)"] = daily_sales.T
weekly_sales
```
